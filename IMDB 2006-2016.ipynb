# Name of the project: IMDB data from 2006 to 2016


Here's a data set of 1,000 most popular movies on IMDB in the last 10 years. The data points included are: Title, Genre, Description, Director, Actors, Year, Runtime, Rating, Votes, Revenue, Metascrore.

Data source: https://www.kaggle.com/PromptCloudHQ/imdb-data

### Data exploration

```python
#Let's begin with the one million dollar question: Which are the top 10 movies between 2006 and 2016? 

films.groupby('Title')[['Genre','Rating']].max().sort_values(by= 'Rating', ascending=False).head(10).reset_index()
```

Drama is the king type above the genres in the top list. 

```python
#Display Title of The Movie Having Runtime Greater Than or equal to 180 Minutes

films[films['Runtime (Minutes)']>=180]['Title']
```

```python
#Display Top 10 Lengthy Movies Title and Runtime

films.nlargest(10, 'Runtime (Minutes)')[['Title', 'Runtime (Minutes)']].set_index('Title')
```

```python
#In Which Year There Was The Highest Average Voting?

films.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

```python
#let's plot the above results

sns.barplot(x='Year', y='Votes', data=films)
plt.title('Votes per Year')
plt.show()
```

The year with the highest number of votes was in 2012 and with the lowest was in 10 years later in 2016.
As we observe there is a huge decrease in the number of votes.
Why is this happening? Were the number of the movies decreased or just the people didn't care much to vote?

```python
#Which was the year with the highest production of movies?

films['Year'].value_counts()
```

There is no correlation between the year and the number of movies. In fact it happeed quite the opposite. Fewer movies were produced the years with the higher number of votes. As the years passing by it's more easy to make movies. But the people it seems that does not care much saying their opinion, at least at this platform. Quality doesn't have to do with quantity.

```python
#Does Rating Affect The Revenue?

sns.scatterplot(x='Rating', y='Revenue (Millions)', data=films)
plt.title('Rating VS Revenue')
plt.show()
```

```python
#Classify Movies Based on Ratings [Excellent, Good, and Average]

def rating(rating):
    if rating >= 7.0:
        return 'Excellent'
    elif rating >= 6.0:
        return 'Good'
    else:
        return 'Average'

films['Rating'].apply(rating)
```

```python
#Add it in a new column in the df

films['Rating_Cat'] = films['Rating'].apply(rating)
films.head()
```

```python
#Create a new df containing only the columns: Title, Rating_Cat

rating_cat = films[['Title','Rating_Cat']]
rating_cat
```

```python
#Find Unique Values From Genre 

films['Genre']
```

```python
#Create a list for all the items

list1 = []
for value in films['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
list2 = []
for item in list1:
    for item1 in item:
       list2.append(item1)
```

```python
list2
```

```python
list3 = []
for item in list2:
    if item not in  list3:
        list3.append(item)
```

```python
list3
```

So there are 20 different genres.

```python
from collections import Counter
final_list = Counter(list2)
df2 = pd.DataFrame.from_dict(final_list, orient='index').reset_index()
df2.head()
```

```python
df2 = df2.rename(columns={'index':'Type', 0:'Count'})
df2
```

```python
df2.groupby('Type')['Count'].max().sort_values(ascending=False).head(10)
```

```python
#Let's do same for the actors

films['Actors']
```

```python
act1 = []
for value in films['Actors']:
    act1.append(value.split(','))
```

```python
act1
```

```python
act2 = []
for value in act1:
    for value1 in value:
        act2.append(value1)    
```

```python
act2
```

```python
act3 = []
for value in act2:
    if value not in act3:
        act3.append(value)
```

```python
act3
```

```python
#In how many films have they played these 10 years?

from collections import Counter
final_act = Counter(act2)
df = pd.DataFrame.from_dict(final_act, orient='index').reset_index()
df
```

```python
df = df.rename(columns={'index':'Name', 0:'Count'})
df
```

```python
#Which actors appear most in this decade?

df.groupby('Name')['Count'].max().sort_values(ascending=False).head(10)
```

```python
df.groupby('Name')['Count'].max().sort_values(ascending=False).tail(10)
```
