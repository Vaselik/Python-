### Total unemployment rate in Europe 2009-2020

Data source: https://ec.europa.eu/eurostat/databrowser/view/TPS00203/default/table

Small amount of data. Manipulate the excel file into a new sheet: Sheet 2.

Delete the blank cells for better results.

There is no amount of data for: Liechtenstein, United Kingdom and Albania, so these cells are deleted.




```python
import pandas as pd
```

```python
une_rate = pd.read_excel(r'C:\Users\Vasiliki\Downloads\TPS002031631892503302.xlsx', sheet_name = 'Sheet 2')
une_rate
```

```python
#Rename the "Unnamed: 0" column to "Country"

une_rate = une_rate.rename(columns={'Unnamed: 0': 'Country'})
une_rate.head()
```

```python
#Locate a specific Country

une_rate.loc[une_rate['Country']== 'Greece']
```

```python
#Take some basic statistical data with describe()

une_rate.describe()
```

Pointing out the "count" row, the number range between 33 to 36 because of the "NaN" values as it can be seen to the next cell.

```python
une_rate.isnull().sum()
```

```python
#Sorting data for the year 2020

une_rate.sort_values('2020')
```

Montenegro had the higher unemployment rate (17.9 %) in 2020.

From the other side, Czechia was doing great with u.r at 2.6 % in the same year.


```python
#Adding new column in the existing dataframe

une_rate['Mean'] = une_rate.mean(axis=1)

une_rate.sort_values('Mean', ascending=False).head()
```

The last 10 years the higher unemployment percentage had the above countries

```python
une_rate.sort_values('Mean', ascending=False).tail()
```

#### Plotting the data and extracting useful information

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (10, 5), dpi=80)

greece = plt.plot(une_rate.iloc[10, 1:13], marker = 'o', linestyle='dotted')
europe = plt.plot(une_rate.iloc[0, 1:13])


plt.title("Greece VS Europe")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate")

plt.legend(["Greece", "Europe"])
```

#### Highlight the difference with a visualize manner

```python
fig = plt.figure(figsize = (10, 5), dpi=80)

norway = plt.plot(une_rate.iloc[30, 1:13])
north_macedonia = plt.plot(une_rate.iloc[33, 1:13])


plt.title("Norway VS N.Macedonia")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate")

plt.legend(["Norway", "N.Macedonia"])
```

#### New DataFrame 

```python
new_rate = une_rate[['Country']]
new_rate['Mean 9-20'] = une_rate.iloc[:, 1:].mean(axis=1)


def categorize(new):
    if new < 6:
        return "Low"
    if new > 12:
        return "High"
    else:
        return "Normal"
      
new_rate["Category"] = new_rate['Mean 9-20'].apply(categorize)

new_rate['Mean 9-14'] = une_rate.iloc[:, 1:7].mean(axis=1)
new_rate['Mean 15-20'] = une_rate.iloc[:, 7:].mean(axis=1)

new_rate
```

```python
fig = plt.figure(figsize = (15, 7.5), dpi=80)

plt.xlabel('Countries')

plt.ylabel('Mean Rate')

plt.title('EU Unemployment 2009 - 2020')

plt.bar(new_rate['Country'], new_rate['Mean 9-20'], width = 0.6)

plt.xticks(rotation='vertical')

plt.show()
```
