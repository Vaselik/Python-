```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://www.meteo.gr/lab/book/index.cfm?city_id=31')
page
```

```python
#page.content

#Making our content more classified - well organized, comparing with printing just the "page.content"
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
#To continue, we are going to the URL
#inspect the page and looking for the "dayblock"
#the term that we want to search

soup.find_all(class_ = 'dayblock')

#We differentiate the class_ , showing that we want the class from BeautifulSoup
#Class is also a keyword in python
```

```python
#Making a loop for the specific search
#Getting the html

for day in soup.find_all(class_ = 'dayblock'):
    day.find('subheader_calendar')
    print(day)
    print('*'*80)  #to seperate the registers
```

```python
#We want to take the text from this

for day in soup.find_all(class_ = 'dayblock'):
    day_text = day.find(class_ = 'subheader_calendar').get_text()
    print(day_text)
    print('*'*80)
```

```python
#For each block we execute the following

for block in soup.find_all(class_ = 'dayblock'):
    day = block.find(class_ = 'subheader_calendar')
    dom = block.find(class_ = 'datenumber_calendar').get_text()
    month = block.find(class_ = 'month_calendar').get_text()
    low = block.find(class_ = 'lowtemp').get_text()
    high = block.find(class_ = 'hightemp').get_text()
    
    print(dom, month, low, high)
    print('*'*80)     
    
```

```python
#Create lists
#Import pandas

import pandas as pd

days = []
temperatures = []

for block in soup.find_all(class_ = 'dayblock'):
    day = block.find(class_ = 'subheader_calendar')
    dom = block.find(class_ = 'datenumber_calendar').get_text()
    month = block.find(class_ = 'month_calendar').get_text()
    minmax = block.find(class_ = 'minmax').get_text()
    daysnew = f"{dom} {month}"   #I want to create a list with strings
    tempnew = f"{low} {high}"
    days.append(daysnew)    
    temperatures.append(tempnew)

print(days)
print(temperatures)

#So I created 2 lists for each part
```

```python
#Create dictionary

weather_table = pd.DataFrame({
    'Days': days,
    'Temperatures': temperatures    
})

weather_table
```

```python
#Save it as a csv file or excel

weather_table.to_csv('Ioannina-Week-Weather.csv')
```
